## Load Data

```julia
dpath = "D:\\OneDrive - imec\\GAD2 - Paper\\Manuscript\\ScienceAdvances\\Revision\\Count_Starter\\"  
stcoord = CSV.read(dpath * "StarterCoordTable.csv", DataFrame)
stdist = CSV.read(dpath * "StarterDistanceTable.csv", DataFrame)
DD = matread(dpath * "Coord_starter.mat");
DD = matread(dpath * "listInfoStarter.mat")
```

```julia
DD["exp"]
```

```julia
df = DataFrame()
dd = []
aa = []
for i = 1:6
    if i == 1
        dd = 1000 .- DD["exp"][2,i][:]
        aa = (.4 * i) .+ zeros(length(dd))

    else
        append!(dd, 1000 .- DD["exp"][2,i][:])
        append!(aa, (.4 * i) .+ zeros(length(DD["exp"][2,i][:])))
    end

end
# Add Depths
df.Depth=dd
# Add Animal
df.Animal = aa
df;
```
